# 1. Run frontend riêng Docker: 

Tôi sẽ hướng dẫn bạn cách chạy Dockerfile này. Đây là các bước cần thực hiện:

1. **Build Docker image**
```bash
docker build -t frontend-app ./frontend
```
- `-t frontend-app`: đặt tên cho image là "frontend-app"
- `./frontend`: đường dẫn đến thư mục chứa Dockerfile

2. **Chạy container**
```bash
docker run -p 5000:5000 frontend-app
```
- `-p 5000:5000`: map port 5000 của container với port 5000 của máy host
- `frontend-app`: tên image đã build ở bước trước

**Lưu ý quan trọng:**
- Đảm bảo bạn đang ở thư mục cha của thư mục `frontend` khi chạy lệnh build
- Đảm bảo file `package.json` của bạn có script `start` được định nghĩa đúng
- Port 5000 phải khớp với port mà ứng dụng Node.js của bạn đang lắng nghe

Nếu bạn muốn chạy container trong background (chế độ detached):
```bash
docker run -d -p 5000:5000 frontend-app
```

Để xem logs của container:
```bash
docker logs <container_id>
```

Tôi sẽ giải thích vấn đề và cách khắc phục:

1. **Nguyên nhân lỗi:**
- Server Node.js của bạn (`server.js`) đang không xử lý đúng các route khi có query parameters (phần sau dấu ? trong URL)
- Khi chạy với live-server (port 5500), nó tự động xử lý routing, nhưng với Node.js server tự viết thì cần cấu hình thêm

2. **Cách sửa `server.js`:**

````javascript:frontend/server.js
const http = require('http');
const fs = require('fs');
const path = require('path');

const server = http.createServer((req, res) => {
  // Lấy đường dẫn cơ bản (bỏ query parameters)
  let filePath = '.' + req.url.split('?')[0];
  
  // Xử lý đường dẫn mặc định
  if (filePath === './') {
    filePath = './index.html';
  }

  // Xử lý các route chính
  if (filePath === './index.html' || filePath === './UIBuilder/index.html') {
    filePath = path.join(__dirname, filePath);
  }

  const extname = String(path.extname(filePath)).toLowerCase();
  const contentType = {
    '.html': 'text/html',
    '.js': 'text/javascript',
    '.css': 'text/css',
    '.json': 'application/json',
    '.png': 'image/png',
    '.jpg': 'image/jpg',
  };

  fs.readFile(filePath, (error, content) => {
    if (error) {
      if (error.code === 'ENOENT') {
        // Nếu không tìm thấy file, thử load index.html
        fs.readFile(path.join(__dirname, 'index.html'), (err, content) => {
          if (err) {
            res.writeHead(404);
            res.end('File not found');
          } else {
            res.writeHead(200, { 'Content-Type': 'text/html' });
            res.end(content, 'utf-8');
          }
        });
      } else {
        res.writeHead(500);
        res.end('Server error: ' + error.code);
      }
    } else {
      res.writeHead(200, { 'Content-Type': contentType[extname] || 'text/plain' });
      res.end(content, 'utf-8');
    }
  });
});

const port = process.env.PORT || 5000;
server.listen(port, () => {
  console.log(`Server running at port ${port}`);
});
````

3. **Giải thích các thay đổi:**
- Tách query parameters khỏi đường dẫn bằng `req.url.split('?')[0]`
- Thêm xử lý cho các route chính
- Nếu không tìm thấy file, thử load `index.html` (Single Page Application behavior)
- Thêm nhiều content types hơn

4. **Cập nhật Dockerfile frontend:**

````dockerfile:frontend/Dockerfile
FROM node:14

WORKDIR /app

# Copy package files
COPY package*.json ./

# Install dependencies
RUN npm install

# Copy all remaining files
COPY . .

# Đảm bảo các file có quyền thực thi
RUN chmod +x /app/server.js

EXPOSE 5000

# Sử dụng nodemon để auto-reload trong development
CMD ["node", "server.js"]
````

5. **Kiểm tra cấu trúc thư mục:**
Đảm bảo cấu trúc thư mục trong container giống với local:
```
frontend/
  ├── index.html
  ├── server.js
  ├── UIBuilder/
  │   └── index.html
  └── ...
```

Những thay đổi này sẽ giúp:
- Xử lý đúng các route có query parameters
- Phục vụ static files đúng cách
- Hỗ trợ Single Page Application routing
- Giữ nguyên trạng thái khi refresh trang

Sau khi cập nhật, hãy rebuild và chạy lại container:
